```python
def fbref_players_data(team):
    players_data = []
    team_name = team['name']
    team_url = team['href']
    
    # get player links
    time.sleep(5)
    response = requests.get(team_url)
    html_content = response.content
    soup = bs4.BeautifulSoup(html_content, 'html.parser')
    table = soup.find('table')
    a_tags = table.find_all('a')
    # filtering for player tags only     
    a_tags = [a for a in a_tags if '/en/players' in a.get('href', '') and not re.search(r'/en/players/[a-z0-9]+/matchlogs', a.get('href', ''))]

    for a_tag in a_tags:
        print(a_tag.get('href'))
    print(len(a_tags))
    
    # get player list names to loop over
    player_table = pd.read_html(team_url)
    players_df = player_table[0]
    players_df.columns = players_df.columns.droplevel(0)
    players_df = players_df[players_df['Min'] > 0]
    players = players_df['Player'].tolist()[:-2]
    print(players)
    print(len(players))
    
    for idx, player in enumerate(players):
        player_dict = {}
        player_dict['name'] = player
        player_dict['position'] = players_df.loc[idx]['Pos']
        player_dict['age'] = int(players_df.loc[idx]['Age'].split('-')[0])
        player_dict['minutes'] = int(players_df.loc[idx]['Min'])
        # get player per ninety and percentile data
        player_href = a_tags[idx].get('href')
        player_dict = get_player_percentile_data(player_dict, str(player_href))
        players_data.append(player_dict)
        
    return players_data
```

```python
def get_player_percentile_data(player_dict, player_link, url='https://fbref.com/'):
    url += player_link
    print(url)
    time.sleep(5)
    res = requests.get(url)
    soup = bs4.BeautifulSoup(res.content, 'html.parser')
    
    # player image
    try:
        img_tag = soup.find('div', attrs={'class' : 'media-item'}).find('img')
        image_url = img_tag["src"]
        player_dict['image_url'] = image_url
    except:
        player_dict['image_url'] = ""
    
    # player per ninety and percentile stats
    try:
        player_tables = pd.read_html(url)
        player_df = player_tables[0]
        player_df = player_df.dropna()
        player_dict['Per 90'] = dict(zip(player_df['Statistic'], player_df['Per 90']))
        player_dict['Percentile'] = dict(zip(player_df['Statistic'], player_df['Percentile']))
    except:
        player_dict['Per 90'] = {}
        player_dict['Percentile'] = {}
        
    print(player_dict)
    
    return player_dict
```

```python
def pass_network(pass_data):
    df = pass_data.copy()
    df['passer'] = df['playerId']
    df['recipient'] = df['passer'].shift(-1)
    passes = df[df['type_displayName'] == 'Pass']
    successful = passes[passes['outcomeType_displayName'] == 'Successful']
    subs = df[df['type_displayName'] == 'SubstitutionOff']
    subs = subs['minute']
    firstSub = subs.min()
    successful = successful[successful['minute'] < firstSub]
    shirt_no = pd.to_numeric(successful['shirtNo'], downcast='integer')
    pas = pd.to_numeric(successful['passer'],downcast='integer')
    rec = pd.to_numeric(successful['recipient'],downcast='integer')
    successful['shirtNo'] = shirt_no
    successful['passer'] = pas
    successful['recipient'] = rec
    player_jersey = successful.groupby(['passer', 'shirtNo'])['x'].mean().reset_index()
    player_shirt_mapping = {i:k for i, k in zip(player_jersey['passer'], player_jersey['shirtNo'])}
    player_name = successful.groupby(['passer', 'name'])['x'].mean().reset_index()
    player_name_mapping = {i:k for i, k in zip(player_name['passer'], player_name['name'])}
        
    average_locations = successful.groupby('passer').agg({'x':['mean'], 'y':['mean', 'count']})
    average_locations.columns = ['x', 'y', 'count']
    df_copy = average_locations.copy()
    df_copy.reset_index(inplace=True)
#     df_copy['shirtNo'] = df_copy['passer'].map(player_mapping)
    pass_between = successful.groupby(['passer', 'recipient']).id.count().reset_index()
    pass_between.rename({'id':'pass_count'}, axis='columns', inplace=True)
    pass_between = pass_between.merge(average_locations, left_on='passer', right_index=True)
    pass_between = pass_between.merge(average_locations, left_on='recipient', right_index=True, suffixes=['', '_end'])
    pass_between = pass_between[pass_between['pass_count'] > 3]
    average_locations.reset_index(inplace=True)
    average_locations['shirtNo'] = average_locations['passer'].map(player_shirt_mapping)
    average_locations['name'] = average_locations['passer'].map(player_name_mapping)
    return (average_locations, pass_between)
    
    #plotting
#     arrows = pitch.arrows(pass_between.x, pass_between.y, pass_between.x_end, pass_between.y_end,
#                           ax=ax, width=5,headwidth=3,color='blue',zorder=1,alpha=0.8)
#     nodes = pitch.scatter(average_locations.x, average_locations.y,
#                           s=1000,color='#FF0000',edgecolors='black',linewidth=5,alpha=1,zorder=2,ax=ax)
        
#     for x in range(len(df_copy['passer'])):
#         pitch.annotate(text='{}'.format(df_copy['shirtNo'][x]), xy=(df_copy['x'][x], df_copy['y'][x]),
#                        ha='center', va='center', c='white', ax=ax, size=10)
        
# #     ax.set_title('Vs {}'.format(opp_team), fontsize=30, color="#082630",
# #                              fontfamily = "Century Gothic", fontweight = "bold", va="center", ha="center", pad=15)
#     plt.savefig('pass_network_twitter_bot.png', dpi=300, bbox_inches='tight')
```
